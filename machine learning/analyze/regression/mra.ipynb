### 범죄 데이터 전처리 완료

```python
month_crime = pd.read_csv('crime_seoul_all.csv', encoding='cp949', index_col=0)
month_crime.head()
```

```python
weather.head()
```

```python
weather.columns
```

```python
df = pd.merge(month_crime, weather, on='일시')
```

```python
# 다변량 회귀 분석 모델 생성
X = df[['강수량(mm)', '일최다강수량(mm)', '평균기온(℃)', '평균최고기온(℃)', '최고기온(℃)',
       '평균최저기온(℃)', '최저기온(℃)', '평균풍속(m/s)', '최대풍속(m/s)', '최대순간풍속(m/s)',
       '평균습도(%rh)', '최저습도(%rh)']] # 독립 변수
y = df['월_범죄건수'] # 종속 변수

X = sm.add_constant(X) # 상수항 추가

model = sm.OLS(y, X).fit() # 최소자승법(OLS)을 사용하여 모델 학습

# 회귀 분석 결과 출력
print(model.summary())
```

```python
# 다변량 회귀 분석 모델 생성
X = df[['강수량(mm)', '일최다강수량(mm)', '평균기온(℃)', '평균최고기온(℃)', '최고기온(℃)',
       '평균최저기온(℃)', '최저기온(℃)', '평균풍속(m/s)', '최대풍속(m/s)', '최대순간풍속(m/s)',
       '평균습도(%rh)', '최저습도(%rh)']] # 독립 변수
y = df['월_범죄건수'] # 종속 변수

X = sm.add_constant(X) # 상수항 추가

model = sm.OLS(y, X).fit() # 최소자승법(OLS)을 사용하여 모델 학습

# 회귀 분석 결과 출력
print(model.summary())
```
