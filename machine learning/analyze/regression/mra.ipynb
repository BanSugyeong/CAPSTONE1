```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 독립 변수와 종속 변수 설정
X = pop[['2020_30대_남자', '2021_30대_남자', '2022_30대_남자', '2020_40대_여자', '2021_40대_여자', '2022_40대_여자']]  # 독립 변수들의 열을 지정합니다.
y = pop['총범죄건수']  # 종속 변수를 지정합니다.
```

```python
# 훈련 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 회귀 모델 생성
model = RandomForestRegressor(n_estimators=100, random_state=42)

# 모델 학습
model.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred = model.predict(X_test)

# 평가 지표 계산 (예: 평균 제곱 오차)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# MAE 계산
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", rmse)
```

### RandomForest 모델 성능 향상

```python
# 이상치 식별 및 처리 함수 정의
def remove_outliers(df, column, threshold=3):
    mean = np.mean(df[column])
    std = np.std(df[column])
    outliers = df[(df[column] - mean) / std > threshold]
    df = df.drop(outliers.index)
    return df

# 이상치 식별 및 처리
threshold = 3  # 임계값 설정
column = '총범죄건수'  # 대상 칼럼 설정
pop_no_outliers = remove_outliers(pop, column, threshold)

# 처리 후 데이터 확인
print("이상치 처리 전 데이터 개수:", len(pop))
print("이상치 처리 후 데이터 개수:", len(pop_no_outliers))
```

```python
# 독립 변수와 종속 변수 설정
X = pop_no_outliers[['2020_30대_남자', '2021_30대_남자', '2022_30대_남자', '2020_40대_여자', '2021_40대_여자', '2022_40대_여자']]  # 독립 변수들의 열을 지정합니다.
y = pop_no_outliers['총범죄건수']  # 종속 변수를 지정합니다.

# 훈련 데이터와 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 회귀 모델 생성
model = RandomForestRegressor(n_estimators=100, random_state=42)

# 모델 학습
model.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred = model.predict(X_test)

# 평가 지표 계산 (예: 평균 제곱 오차)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# MAE 계산
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor

# 각각의 기본 모델 설정
model1 = RandomForestRegressor(n_estimators=100, random_state=42)
model2 = RandomForestRegressor(n_estimators=200, random_state=42)
model3 = RandomForestRegressor(n_estimators=300, random_state=42)

# VotingRegressor 생성
ensemble_model_rf = VotingRegressor(estimators=[('rf1', model1), ('rf2', model2), ('rf3', model3)])

# 모델 학습
ensemble_model_rf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred_rf = ensemble_model_rf.predict(X_test)

# 평가 지표 계산 (예: 평균 제곱 오차)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print("Mean Squared Error (Random Forest Ensemble):", rmse_rf)
print("예측 결과 : ", y_pred_rf)
```
